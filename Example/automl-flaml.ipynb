```python
import orchest
# Retrieve the data from the previous step.
data = orchest.get_inputs()
train, test = data["clean_data"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = train.drop(columns = 'Class')
y = train.loc[:,'Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from flaml import AutoML
# Initialize an AutoML instance
automl = AutoML()
# Specify automl goal and constrainta
automl_settings = {
    "time_budget": 240,  # total running time in seconds
    "metric": 'accuracy',  # can be: 'r2', 'rmse', 'mae', 'mse', 'accuracy', 'roc_auc', 'roc_auc_ovr',
                           # 'roc_auc_ovo', 'log_loss', 'mape', 'f1', 'ap', 'ndcg', 'micro_f1', 'macro_f1'
    "task": 'classification',  # task type    
    "log_file_name": 'titanic.log',  # flaml log file
    "seed": 1429,    # random seed
}
```

```python
# Train with labeled input data
automl.fit(X_train=X_train, y_train=y_train,
           **automl_settings)
```

```python
# Predict
y_pred = automl.predict(X_test)
y_pred_proba = automl.predict_proba(X_test)[:,1]

```

```python
''' retrieve best config and best learner'''
print('Best ML leaner:', automl.best_estimator)
print('Best hyperparmeter config:', automl.best_config)
print('Best accuracy on validation data: {0:.4g}'.format(1-automl.best_loss))
print('Training duration of best run: {0:.4g} s'.format(automl.best_config_train_time))
```

```python
# Export the best model
print(automl.model)
```

```python
automl.model.estimator
```

```python
''' compute different metric values on testing dataset'''
from flaml.ml import sklearn_metric_loss_score
print('accuracy', '=', 1 - sklearn_metric_loss_score('accuracy', y_pred, y_test))
print('roc_auc', '=', 1 - sklearn_metric_loss_score('roc_auc', y_pred_proba, y_test))
print('log_loss', '=', sklearn_metric_loss_score('log_loss', y_pred_proba, y_test))
```

```python

''' pickle and save the automl object '''
import pickle
with open('automl.pkl', 'wb') as f:
    pickle.dump(automl, f)
```
